```python
import pandas as pd
```

```python
data = pd.read_csv(r'DSBDALExam DataSets\Adult\adult.csv', header=None)
```

```python
data.head()
```

```python
data.columns
```

```python
# Set correct column names
columns = [
    'age', 'workclass', 'fnlwgt', 'education', 'education-num', 'marital-status',
    'occupation', 'relationship', 'race', 'sex', 'capital-gain', 'capital-loss',
    'hours-per-week', 'native-country', 'income'
]
data.columns = columns
```

```python
data.head()
```

```python
# Create data subsets for different Country, Sex, race.
print(data['native-country'].value_counts())
print("\n", data['sex'].value_counts())
print("\n", data['race'].value_counts())

native_india = data[data['native-country'] == 'India']
female_data = data[data['sex'] == 'Female']
male_data = data[data['sex'] == 'Male']
black_data = data[data['race'] == 'Black']
```

```python
# Merge two subsets [inner merge (intersection)]
merged_data = pd.merge(female_data, black_data, how='inner')

print(merged_data.head())
```

```python
# Sort Data using customer ratings.
sorted_data = data.sort_values(by='hours-per-week', ascending=False)

print(sorted_data)
```

```python
# Transposing Data
transposed_data = data.transpose()

print(transposed_data.head())
```

```python
# Melting Data to long format  
melted = pd.melt(data,
                 id_vars=['age', 'sex'],
                 value_vars=['education', 'occupation'],
                 var_name='Attribute',
                 value_name='Value')
print(melted.head())
```

```python
print(melted.columns)
```

```python
# Casting data to wide format 
wide_data = melted.pivot_table(index=['age', 'sex'],
                                   columns='Attribute',
                                   values='Value',
                                   aggfunc='first').reset_index()
print(wide_data.head())
```
