```python
# Create data subsets for different languages(Original Language).  
english_movies = movies[movies['original_language'] == 'en']
french_movies = movies[movies['original_language'] == 'fr']
italian_movies = movies[movies['original_language'] == 'it']
```

```python
# Merge two subsets
merged_movies = pd.concat([english_movies, french_movies], ignore_index=True)
print(merged_movies.head())
print(merged_movies.tail())
```

```python
# Sort Data using customer ratings
sorted_data = movies.sort_values(by='vote_average', ascending=False)
print(sorted_data.head())
```

```python
# Transposing Data
transposed_movies = movies.head().transpose()
print(transposed_movies.head())
```

```python
# Melting Data to long format
melted_movies = pd.melt(movies[['id', 'original_language', 'title']], id_vars=['id'], 
                        var_name='Attribute', value_name='Value')
print(melted_movies.head())
```

```python
# Casting data back to wide format
wide_movies = melted_movies.pivot_table(index='id', 
                                        columns='Attribute', 
                                        values='Value', 
                                        aggfunc='first').reset_index()

print(wide_movies.head())
```
