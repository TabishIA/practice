```python
'''
Perform the following operations using Python on the Facebook metrics data sets  
a. Create data subsets for type of post  
b. Merge two subsets  
c. Sort Data on Page total likes  
d. Transposing Data  
e. Melting Data to long format  
f. Casting data to wide format 
'''
```

```python
import pandas as pd
```

```python
data = pd.read_csv('DSBDALExam DataSets\dataset_Facebook.csv', delimiter=';')
```

```python
data.head()
```

```python
print(data['Type'].value_counts())
```

```python
# Create data subsets for type of posts
photo_posts = data[data['Type'] == 'Photo']
status_posts = data[data['Type'] == 'Status']
link_posts = data[data['Type'] == 'Link']
video_posts = data[data['Type'] == 'Video']
```

```python
# Merge two data subsets
merged_data = pd.concat([photo_posts, status_posts])

print("\nMerged Data (Photo + Status Posts):")
print(merged_data.head())
```

```python
# Sort Data on Page total likes
sorted_data = data.sort_values(by='Page total likes', ascending=False)
print("\nData Sorted by Page Total Likes:")
print(sorted_data.head())
```

```python
# Transposing Data
transposed_data = data.head().T # Transpose first 5 rows for display
print("\nTransposed Data:")
print(transposed_data)
```

```python
# Melting Data to long format
melted_data = pd.melt(data, id_vars=['Type', 'Category'],
                      value_vars=['like', 'share', 'comment'])
print("\nMelted Data (Long Format):")
print(melted_data.head())
```

```python
# Casting data to wide format (pivoting melted data)
wide_data = melted_data.pivot_table(index=['Type', 'Category'], 
                                    columns='variable', values='value', aggfunc='mean')
print("\nWide Format Data (Pivot Table):")
print(wide_data.head())
```
